```python
from PIL import Image
from IPython.display import display
im = Image.open('protoniclogo.png')
display(im)
# we have imported the image with the use of PIL library and displayed with the help of IPython.display library
```

```python
# now we will convert thisn image into array format
array6 = np.array(im)
print(array6.shape)
array6
```

here "uint" means unsigned integer means there is no negative value is included in the set of values. 8 means it it 8 bit per byte. this means value can be upto 2*2*2*2*2*2*2*2 = 256 (<=255). for example black is stored as 0 and white is stored as white.

```python
#now we will try to invert the image.
mask = np.full(array6.shape,255)
mask
```

```python
# substract that from modefied array
modefied_array = array6-mask

# no we have to convert all the negative values into positive values.
modefied_array = modefied_array*(-1)

#now last step is to set datatypw correctly with the help of numpy
modefied_array = modefied_array.astype(np.uint8)

modefied_array
```

```python
#we have used fromarray() function to create image from the array.
display(Image.fromarray(modefied_array))
```
