```python
# if we want to use the first column as our index
df = pd.read_csv('datasets/Admission_Predict.csv', index_col=0)
df.head()
```

```python
# we can rename the column names with .rename
new_df = df.rename(columns = {"SOP" : "Statement of purpose", "LOR" : "Letter of Recommandation"})
new_df.head()
```

```python
# in above dataframe the lOR is not renamed because there should be some problem with the name
# let us check
df.columns
```

```python
# we can see there is space after "LOR" and "Chance of Admit"
# we need to remove the spaces from all the column name
new_df = new_df.rename(mapper = str.strip, axis = 'columns')
new_df.columns
```

```python
# now we have removed the extra spaces from the column names
# we can change the name of the LOR
new_df = new_df.rename(columns = {"LOR" : "Letter of Recommandation"})
new_df.head()
```
