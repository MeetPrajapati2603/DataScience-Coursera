```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/presidents.csv')
df.head()
```

```python
df.head()
```

```python
df.columns
```

```python
# here we will create the function for splitting the full name into first and last.
def splitname(row):
    row["First"] = row["President "].split(" ")[0]
    row["Last"] = row["President "].split(" ")[-1]
    return row
df = df.apply(splitname, axis='columns')
df.head()
```

```python
# we will extract the name of presidents
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>^[\w]*$)"

# now we will extract the data with the help of the pattern
names = df["President "].str.extract(pattern).head()
names
```

```python
df["Took office "] = pd.to_datetime(df["Took office "])
df["Took office "].head()
```
